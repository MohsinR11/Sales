### **üì¶ Import Libraries & Load Data**

```python
# üì¶ Importing essential libraries for data analysis and visualization
import pandas as pd                # For data manipulation and analysis
import numpy as np                 # For numerical operations (optional in this snippet)
import seaborn as sns              # For advanced data visualizations
import matplotlib.pyplot as plt    # For basic plotting

# üì• Load the sales dataset from a CSV file into a DataFrame
df = pd.read_csv("sales.csv")     # Replace "sales.csv" with the full file path if needed

# üïí Convert the 'Date' column to datetime format to enable time-based operations
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# üëÄ Display the first 5 rows of the dataset to get a quick overview
df.head()
```

---
---

### **üìã Basic Info & Preprocessing**

```python
# üìä OVERVIEW OF THE DATAFRAME

# üîé Check the structure and data types of each column
print(df.info())  # Shows non-null counts, data types, and memory usage

# üìà Get basic statistical details for numeric columns
print(df.describe())  # Includes mean, std, min, max, etc.

# üßπ Check for missing values in each column
print(df.isnull().sum())  # Helps identify if any column has null/missing data


# üßΩ DATA CLEANING

# üîÅ Remove duplicate rows, if any, to ensure data quality
df.drop_duplicates(inplace=True)


# üìÖ FEATURE ENGINEERING ‚Äî Extracting useful time-based features

# üìÜ Extract year from the 'Date' column
df['Year'] = df['Date'].dt.year

# üóìÔ∏è Extract full month name (e.g., January, February)
df['Month'] = df['Date'].dt.month_name()

# üî¢ Extract day of the month
df['Day'] = df['Date'].dt.day

# üìÖ Extract the weekday name (e.g., Monday, Tuesday)
df['Weekday'] = df['Date'].dt.day_name()
```

### **üìä Univariate Analysis**

#### **üîπ Revenue Distribution**

```python
# üìä DATA VISUALIZATION - REVENUE DISTRIBUTION

# üñºÔ∏è Set the figure size for the plot (width: 11, height: 5)
plt.figure(figsize=(11, 5))

# üìà Create a histogram with a KDE (Kernel Density Estimate) for the 'Revenue' column
sns.histplot(df['Revenue'], color='deepskyblue', kde=True)

# üè∑Ô∏è Set the title of the plot to describe what we're visualizing
plt.title("Revenue Distribution")

# üö´ Disable the grid lines for a cleaner plot
plt.grid(False)

# üñºÔ∏è Display the plot
plt.show()
```

```python
# üìä DATA VISUALIZATION - PROFIT DISTRIBUTION

# üñºÔ∏è Set the figure size for the plot (width: 11, height: 5)
plt.figure(figsize=(11, 5))

# üìà Create a histogram with a KDE (Kernel Density Estimate) for the 'Profit' column
sns.histplot(df['Profit'], kde=True, color='lime')

# üè∑Ô∏è Set the title of the plot to describe what we're visualizing
plt.title("Profit Distribution")

# üö´ Disable the grid lines for a cleaner plot
plt.grid(False)

# üñºÔ∏è Display the plot
plt.show()
```

### **üìà 4. Bivariate Analysis (Correlations)**

#### **üîπ Revenue vs Profit**

```python
# üìä DATA VISUALIZATION - REVENUE vs PROFIT (BY CATEGORY)

# üñºÔ∏è Set the figure size for the plot (width: 11, height: 5)
plt.figure(figsize=(11, 5))

# üîµ Create a scatter plot to show the relationship between 'Revenue' and 'Profit'
# 'hue' is used to color points by 'Category' and 's' adjusts the size of the points
sns.scatterplot(x='Revenue', y='Profit', data=df, hue='Category', s=140)

# üè∑Ô∏è Set the title of the plot to describe what we're visualizing
plt.title("Revenue vs Profit (by Category)")

# üö´ Disable the grid lines for a cleaner plot
plt.grid(False)

# üñºÔ∏è Display the plot
plt.show()
```

---

#### **üîπ Correlation Matrix**

```python
# üìä DATA VISUALIZATION - CORRELATION HEATMAP

# üîç Calculate the correlation matrix for the 'Revenue', 'Profit', and 'Quantity' columns
corr = df[['Revenue', 'Profit', 'Quantity']].corr()

# üñºÔ∏è Set the figure size for the plot (width: 11, height: 5)
plt.figure(figsize=(11, 5))

# üå°Ô∏è Create a heatmap to visualize the correlation between the selected columns
# 'annot=True' displays the correlation values on the heatmap, and 'cmap' defines the color scheme
sns.heatmap(corr, annot=True, cmap='PuRd')

# üè∑Ô∏è Set the title of the heatmap to describe what we're visualizing
plt.title("Correlation Heatmap")

# üñºÔ∏è Display the heatmap plot
plt.show()
```

### **üï∞Ô∏è Time Series Trends**

#### **üîπ Monthly Revenue Trend**

```python
# üìä DATA VISUALIZATION - MONTHLY REVENUE TREND

# üî¢ Group the data by 'Year' and 'Month' to calculate the total revenue for each month
# Use 'sum()' to aggregate the revenue per group, and reset the index to create a clean DataFrame
monthly_rev = df.groupby(['Year', 'Month'])['Revenue'].sum().reset_index()

# üìÖ Create a 'Date' column by combining the 'Year' and 'Month' columns and converting it to datetime format
monthly_rev['Date'] = pd.to_datetime(monthly_rev['Year'].astype(str) + '-' + monthly_rev['Month'].astype(str))

# üñºÔ∏è Set the figure size for the plot (width: 11, height: 5)
plt.figure(figsize=(11, 5))

# üìà Create a line plot to visualize the monthly revenue trend over time
# 'sort_values' ensures the dates are ordered correctly on the x-axis
sns.lineplot(data=monthly_rev.sort_values('Date'), x='Date', y='Revenue', color='red')

# üè∑Ô∏è Set the title of the plot to describe what we're visualizing
plt.title("Monthly Revenue Trend")

# üö´ Disable the grid lines for a cleaner plot
plt.grid(False)

# üñºÔ∏è Display the plot
plt.show()
```

---

#### **üîπ Profit by Region**

```python
# üìä DATA VISUALIZATION - TOTAL PROFIT BY REGION

# üñºÔ∏è Set the figure size for the plot (width: 11, height: 5)
plt.figure(figsize=(11, 5))

# üìä Create a bar plot to visualize the total profit by region
# Group data by 'Region' and calculate the sum of 'Profit' for each region using 'groupby()' and 'sum()'
# Reset the index to make the DataFrame clean for plotting
sns.barplot(data=df.groupby('Region')['Profit'].sum().reset_index(),
            x='Region',        # Set 'Region' as the x-axis
            y='Profit',        # Set 'Profit' as the y-axis
            color='lime',      # Set the color of the bars
            width=0.6,         # Adjust the width of the bars
            estimator=sum,     # Use 'sum' as the estimator to aggregate profits per region
            errorbar=None,     # Remove error bars from the plot
            palette='Set2')    # Use the 'Set2' color palette for the bars

# üè∑Ô∏è Set the title of the plot to describe what we're visualizing
plt.title("Total Profit by Region")

# üîÑ Rotate the x-axis labels (region names) for better readability
plt.xticks(rotation=45)

# üö´ Disable the grid lines for a cleaner plot
plt.grid(False)

# üñºÔ∏è Display the plot
plt.show()
```

### **üö® Outlier Detection**

#### **üîπ Revenue & Profit Boxplots**

```python
# üìä DATA VISUALIZATION - REVENUE AND PROFIT OUTLIERS

# üñºÔ∏è Create a figure with 1 row and 2 columns of subplots, with a total figure size of (width: 11, height: 5)
fig, axes = plt.subplots(1, 2, figsize=(11, 5))

# üì¶ Create a boxplot for the 'Revenue' column to visualize outliers
# The boxplot is drawn on the first subplot (axes[0]) and colored red
sns.boxplot(ax=axes[0], y=df['Revenue'], color='red')

# üè∑Ô∏è Set the title for the first subplot (Revenue outliers)
axes[0].set_title("Revenue Outliers")

# üì¶ Create a boxplot for the 'Profit' column to visualize outliers
# The boxplot is drawn on the second subplot (axes[1]) and colored aqua
sns.boxplot(ax=axes[1], y=df['Profit'], color='aqua')

# üè∑Ô∏è Set the title for the second subplot (Profit outliers)
axes[1].set_title("Profit Outliers")

# üîß Adjust layout to avoid overlapping elements
plt.tight_layout()

# üö´ Disable grid lines for both subplots for a cleaner appearance
axes[0].grid(False)
axes[1].grid(False)

# üñºÔ∏è Display the plots
plt.show()
```

## **üí° Key Business Insights**

#### **üìå Top-Selling Categories**

```python
# üìä DATA ANALYSIS - TOP CATEGORIES BY REVENUE

# üî¢ Group the data by 'Category' and calculate the total revenue for each category using 'sum()'
# Then, sort the revenue values in descending order to get the top categories
top_cat = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)

# üìù Print the sorted total revenue by category to view the top categories
print(top_cat)
```

---

#### **üìå High-Profit Products**

```python
# üìä DATA ANALYSIS - TOP 10 PRODUCTS BY PROFIT

# üî¢ Group the data by 'Product' and calculate the total profit for each product using 'sum()'
# Then, sort the profit values in descending order to get the top products
# Use 'head(10)' to select the top 10 products with the highest profit
top_profit = df.groupby('Product')['Profit'].sum().sort_values(ascending=False).head(10)

# üìù Print the top 10 products by total profit
print(top_profit)
```

---

#### **üìå Low/Negative Profit Areas**

```python
# üìä DATA ANALYSIS - REGIONS WITH LOWEST PROFIT

# üî¢ Group the data by 'Region' and calculate the total profit for each region using 'sum()'
# Then, sort the profit values in ascending order to get the regions with the lowest profit
low_profit = df.groupby('Region')['Profit'].sum().sort_values()

# üìù Print the regions with the lowest total profit
print(low_profit)
```
